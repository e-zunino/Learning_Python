```python
obj4 = pd.Series(sdata, index=states)
```

### Reindexing

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=["d", "b", "a", "c"])
```

```python
obj2=obj.reindex(["a","b","c","d","e"])
```

```python
obj2
```

You may want to do some interpolation or filling of values when reindexing. The method option allows us to do this, using a method such as `ffill`, which forward-fills **the values**:

```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
```

```python
obj3
```

```python
obj4=obj3.reindex(index=np.arange(6), method="ffill")
```

### Arithmetic and Data Alignment

pandas can make it much simpler to work with objects that have different indexes. For example, when you add objects, if any index pairs are not the same, the respective index in the result will be the union of the index pairs

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=["a", "c", "e", "f", "g"])
```

```python
s1+s2
```

In the case of DataFrame, alignment is performed on both rows and columns:

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list("bcd"), index=["Ohio", "Texas", "Colorado"])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list("bde"), index=["Utah", "Ohio", "Texas", "Oregon"])
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
df1.index
```

```python
df1.reindex(["Texas", "Ohio", "Colorado"])
```

```python
df1+df2
```

If you add DataFrame objects with no column or row labels in common, the result will contain all nulls. In arithmetic operations between differently indexed objects, you might want to fill with a special value, like 0, when an axis label is found in one object but not the other.

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), columns=list("abcd"))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)), columns=list("abcde"))
```

```python
df1+df2
```

```python
df1.add(df2, fill_value=0)
```

```python
df1.add(df2, fill_value=1000)
```

When reindexing a Series or DataFrame, you can also specify a different fill value:

```python
df1.reindex(columns=df2.columns, fill_value=0)
```

Arithmetic between DataFrame and Series is also defined. Let us consider an example using arrays

```python
arr = np.arange(12.).reshape((3, 4))
```

```python
arr-arr[0]
```

This is an example of **broadcasting**. By default, arithmetic between DataFrame and Series matches the index of the Series on the columns of the DataFrame, broadcasting down the rows:

```python
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns=list("bde"), index=["Utah", "Ohio", "Texas", "Oregon"])
```

```python
series = frame.iloc[0]
```

```python
frame
```

```python
series
```

```python
frame-series
```

```python
series2 = pd.Series(np.arange(3), index=["b", "e", "f"])
```

```python
frame+series2
```

If you want to instead broadcast over the columns, matching on the rows, you have to use one of the arithmetic methods and specify to match over the index.

```python
series3 = frame["d"]
```

```python
frame
```

```python
series3
```

```python
frame.sub(series3, axis="index")
```

### Function Application and Mapping

```python
frame = pd.DataFrame(np.random.standard_normal((4, 3)), columns=list("bde"), index=["Utah", "Ohio", "Texas", "Oregon"])
```

```python
frame
```

```python
np.abs(frame)
```

Another frequent operation is applying a function on one-dimensional arrays to each column or row. DataFrame’s apply method does exactly this:

```python
def f1(x): 
    return x.max() - x.min()
```

```python
frame.apply(f1)
```

If you pass `axis="columns"` to apply, the function will be invoked once per row instead.

```python
frame.apply(f1, axis="columns")
```

The function passed to apply need not return a scalar value; it can also return a Series with multiple values:

```python
def f2(x): 
    return pd.Series([x.min(), x.max()], index=["min", "max"])
```

```python
frame.apply(f2)
```

```python
def my_format(x): 
    return f"{x:.2f}"
```

```python
fr2=frame.map(my_format)
```

```python
fr2["e"].dtype
```

### Sorting and Ranking

```python
obj = pd.Series(np.arange(4), index=["d", "a", "b", "c"])
```

```python
obj.sort_index()
```

```python
obj.sort_values()
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2, 4)), index=["three", "one"], columns=["d", "a", "b", "c"])
```

```python
frame
```

```python
frame.sort_index()
```

```python
frame.sort_index(axis="columns", ascending=True) #default ascending=True
```

Any missing values are sorted to the end of the Series by default

```python
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
```

```python
obj.sort_values()
```

```python
obj.sort_values(na_position="first")
```

When sorting a DataFrame, you can use the data in one or more columns as the sort keys. To do so, pass one or more column names to sort_values:

```python
frame = pd.DataFrame({"b": [4, 7, -3, 2], "a": [0, 1, 0, 1]})
```

```python
frame
```

```python
frame.sort_values("b")
```

```python
frame.sort_values(["a","b"])
```

**Ranking** assigns ranks from one through the number of valid data points in an array,  starting from the lowest value. The rank methods for Series and DataFrame are the place to look; by default, rank breaks ties by assigning each group the mean rank:

```python
obj = pd.Series([7, -5, 7, 4, 1000, 0, 4,-1])
```

```python
obj.rank()
```

```python
obj.rank(method="first")
```

```python
frame = pd.DataFrame({"b": [4.3, 7, -3, 2], "a": [0, 1, 0, 1], "c": [-2, 5, 8, -2.5]})
```

```python
frame
```

```python
frame.rank(axis="columns")
```
